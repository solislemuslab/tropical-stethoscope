Interpret the info of boxes from txt with wav data

```python
from scipy.io import wavfile
import numpy as np
import scipy
import scipy.signal
import soundfile as sf
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import csv
from collections import namedtuple
import datetime
import time
```

Read files and extract info of boxes

```python
# # 101 boxes in this wav file
# wavFile = "../wav/3B_20180607_070500_Dawn_Sunrise_Based.wav"
# labelFile = "../wav/3B_20180607_070500_Dawn_Sunrise_Based.txt"
# identifier = "3B_20180607_070500_Dawn_Sunrise_Based" #used to identify the boxex belongs to the wavFile

# 57 boxes in this wav file
wavFile = "../wav/20190205_063000_Full24H [1.4438 116.9805].wav"
labelFile = "../wav/20190205_063000_Full24H [1.4438 116.9805].txt"
identifier = "20190205_063000_Full24H"
```

```python
audio, rate = sf.read(wavFile)
# # normalize data to have unit energy
# data = data / np.sqrt(np.mean(np.square(np.abs(data))))
# labels = np.genfromtxt(labelFile, delimiter='	')

# extract the info of boxes in wavFile
# header = namedtuple('Selection','View','Channel', 'Begin Time (s)','End Time (s)','Low Freq (Hz)','High Freq (Hz)','Begin Path','Begin Date', 'Delta Time (s)', 'Begin Clock Time', 'Sonotype','Group','Type','Spe','obs')
times = [] 
freqs = []
timeWeight = [60,1,0.1]
with open(labelFile) as file:
    label_reader = csv.reader(file, delimiter='\t')
    for row in label_reader:
        if identifier in row[7]: # file identifier in the index 7
            str = sum([a*b for a,b in zip(timeWeight, map(int,row[10].replace('.',":").split(':')))]) % 1800
            end = sum([a*b for a,b in zip(timeWeight, map(int,row[11].replace('.',":").split(':')))]) % 1800
            times.append([str,end])
#             times.append([float(row[3]), float(row[4])])
            freqs.append([float(row[5]), float(row[6])]) 

print(np.max(times))
print(np.max(freqs))
```

Extrect the start and end times, the low and high frequencies <br>
index 1 - 4 of all labels are corresponds to boxes in wav file "3B_20180607_070500_Dawn_Sunrise_Based.wav"

Ramdomly choose one box and plot its spectrogram

```python
# compute the data for the spectrogram
freq, t, spec = scipy.signal.spectrogram(audio, rate)
```

```python
# randomly choose one box and get the audio of the box
boxIndex = random.randint(0, len(freqs) - 1)
boxT = times[boxIndex]
boxFreq = freqs[boxIndex]

print(boxTime)
print(boxFreq)
```

```python
# get low and high freq index of the box
low_freq = np.argmin(np.abs(freq - boxFreq[0])) - 1
high_freq = np.argmin(np.abs(freq - boxFreq[1])) + 1
# get start and end time of the box
start = np.argmin(np.abs(t - boxTime[0])) -1
end = np.argmin(np.abs(t - boxTime[1])) + 1

boxSpecPlt= spec[low_freq:high_freq, start:end]
boxFreqPlt = freq[low_freq: high_freq]
boxTPlt = t[start:end]
```

```python
# plot the spectrogram
# set up
plt.figure(figsize=(15, 5))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.title("Spectrogram for a random box")

# plot
plt.pcolormesh(boxTPlt, boxFreqPlt, boxSpecPlt)
plt.colorbar()
```
